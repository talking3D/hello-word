```python
import pandas as pd
import numpy as np
np.random.seed(2019)

from sklearn.model_selection import cross_val_score

from sklearn.dummy import DummyClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import scikitplot as skplt

from collections import defaultdict

import helper as h

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('../inputs/titanic.train.csv')
train.sample(7)
```

```python
train.pivot_table(index=['Pclass', 'Sex'], values=['Survived'], aggfunc=np.mean)
```

```python
train['title'] = train['Name'].map(lambda x: x.split(',')[1].split('.')[0].strip())
train['title'].value_counts()
```

```python
top_titles = list(train['title'].value_counts().head(6).keys())
top_titles
```

```python
title_norm = defaultdict(lambda: 'other', dict(zip(top_titles, top_titles)))
train['title_norm'] = train['title'].map(title_norm)
train['title_norm'].value_counts()
```
